- CLAHE equalization for colored image

Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Read image and show it

```python
img = cv2.imread("Lenna.png")
print(img.shape)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_rgb)
```

Apply CLAHE

```python
# make and apply CLAHE
CLAHE = cv2.createCLAHE(clipLimit = 1, tileGridSize = (2, 2))
new_hsv = img_hsv.copy()
new_hsv[:,:,2] = CLAHE.apply(img_hsv[:,:,2])

# convert ro rbg and show
new_rgb = cv2.cvtColor(new_hsv,cv2.COLOR_HSV2RGB)

f, axarr = plt.subplots(3,1,figsize=(12,12))

axarr[0].hist(img_hsv[:,:,2].flatten(),256,[0,255],color = 'red')
axarr[1].hist(new_hsv[:,:,2].flatten(),256,[0,255],color = 'red')
axarr[2].imshow(new_rgb)
plt.show()
```
